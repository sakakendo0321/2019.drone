```python
import glob
import json
import pandas as pd
from IPython.core.display import display
import sys
from pandas.io.json import json_normalize
pd.set_option('display.max_columns', 100)
```

```python
def display_items(items):
    try :
#        print(items)
        data=pd.DataFrame.from_dict(items)
#        data2=pd.io.json.json_normalize(data,record_path='demo')
        data2=json_normalize(data['state'])
        display(data)
        display(data2)
        display(json_normalize(data['demo']))
    except:
        print(sys.exc_info())
def json2df(_data):
    data=pd.DataFrame.from_dict(_data)
    display(data)

def file2json(file_name :str)->dict:
    data={'items':[]}
    with open(file_name,'r') as f:
        while True:
            line=f.readline()
            if not line: break
            #print(type(lines),lines)
            data['items'].append(json.loads(line)['droneState'])
    return data
        
def analysis_txt(files):
    for file in files:
        print(file)
        data=file2json(file)
        json2df(data['items'])
#            display_items(data['items'])
        

txt_files=glob.glob('../log/*.txt')

navdata_txt=glob.glob('../log/*/navdata.txt')
analysis_txt(navdata_txt)
```
